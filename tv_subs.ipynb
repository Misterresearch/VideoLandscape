## Description 

TVE Anywhere is an authenticated digital video service for paying Cable TV customers. However, as with other subscription video services (SVOD) such as Netflix and Hulu, logins are routinely shared. The goal is to find "islands" within our network that represent inappropriate sharing, that may rise to the level of fraud. 

Our data set is relatively large, and consists of almost 83K nodes that were generated in one day of usage for a major media publisher. In this case, identifying groups or subgraphs of nodes in our large network is an imperative, in order to turn our large data set into actionable analytics. Our initial network is transformed into a bipartite network, with the top nodes for IP address and the users login ID as the bottom.

Note: All IP addresses, and logins were anonomized prior to analysis. 

```python
# jupyter notebook --NotebookApp.iopub_data_rate_limit=10000000000
import networkx as nx
import pandas as pd
import urllib
import numpy as np
import matplotlib.pyplot as plt 
import sys
import os
from networkx.algorithms import bipartite as bi



url = "https://raw.githubusercontent.com/Misterresearch/CUNY-Projects/master/tveanon_ip_guid.csv"
tvsubs = pd.read_csv(url, parse_dates=True, names=['IP','ID'], header=0)
print tvsubs
```

```python
B = nx.Graph()
B.add_nodes_from(tvsubs['IP'], bipartite=0)
B.add_nodes_from(tvsubs['ID'], bipartite=1)
B.add_edges_from(zip(tvsubs['IP'], tvsubs['ID']),weight=1)

#B is an unconnected graph
bottom_nodes, top_nodes = bi.sets(B)
top_nodes = set(n for n,d in B.nodes(data=True) if d['bipartite']==0)
bottom_nodes = set(B) - top_nodes

Bip = bi.weighted_projected_graph(B, top_nodes, ratio=False)
Bid = bi.weighted_projected_graph(B, bottom_nodes, ratio=False)
#nx.draw(Bip, node_size=0, node_color='w', edge_color='b', alpha=.2)
#plt.show()
```

## Bipartite & Projected Netwroks, Setting Weights for Edges & Nodes

Our original bipartite network (B) is created by using the networkx Graph function, but also with the "add_nodes_from" function with the "bipartite" argument.  Once we've created an initial bipartite network (B), two projected networks are created - on for IP (Bip) and the other for our login ID (Bid). 

It's also important to note that because we're developing subgraphs using the island_method from a projected graph instead of a simple graph, using the "weight" setting in the add_edges_from function, does not work. It's actually the ratio argument (set to false), that gives us an interger weight setting for the edges in our projected graph. From here we can use the custom function "trim_edges" found in Tsvetovat & Kouznetsov's Social Networks for Startups.

## Water Levels - Node Centrality vs Edges

Although we're going to use edge weight to develop our subgraphs, the table above provides a ranking of our projected graph 'IP' nodes based on degree centrality counts, which could also be a viable option. Again, for this specific analysis we've chosen to develop our subgraphs using edge weights instead.

```python
Bip2=Bip.copy()
def trim_edges(Bip, weight=1):
    Bip2=nx.Graph()
    for f,to,edata in Bip.edges(data=True):
                if edata['weight'] > weight:
                        Bip2.add_edge(f,to,edata)
    return Bip2
trim_edges(Bip2, 2)
#E.edges()
Bip2.edges(data=True)[0:10]
```

```python
Bip3=Bip.copy()
def trim_nodes(Bip, weight=1):
    #Bip3=nx.Graph()
    for i,j, in Bip.degree(weight='weight').items():
                if j > weight:
                        Bip3.add_node(i)
    return Bip3
#trim_nodes(Bip3, 1)
Bip3.degree(weight='weight').items()[:10]
```

```python
def island_method(Bip, iterations=3):
    weights= [edata['weight'] for f,to,edata in Bip.edges(data=True)]
    mn=int(min(weights))
    mx=int(max(weights))
    #compute the size of the step, so we get a reasonable step in iterations
    step=int((mx-mn)/iterations)
    return [[threshold, trim_edges(Bip, threshold)] for threshold in range(mn,mx,step)]
island_method(Bip, 3)
```

## Insights & Analysis

As is the case with other fraud detection models, the overall number of nodes of IP addresses, that share a login with another IP address is only 119 - less than 1% of all users for our sample day's worth of usage. We see that there are 35 such "islands" or subgroups in our network. However, this water level of '1", is probably "good" sharing meaning one other "location" could be a paying customer's place of work or even another wifi network in the same household. 

A better threshold might be 2 other IP addresses, this is more likely to be within in the realm of suspicious behavior - and at this level there are 44 nodes across 14 islands that represent sharing at this level. At the opposite extreme is a very small group of nodes and islands with 5 IP address (1 + 4 other IPs), sharing a single login. However, at this level we risk missing fraud at lower "water levels".
